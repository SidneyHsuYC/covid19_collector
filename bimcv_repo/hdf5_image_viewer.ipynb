```python
import h5py
import math
import pandas as pd
import matplotlib.pyplot as plt

from datetime import date
```

```python
covid19_pandas = 'bimcv_metadata.csv'
covid19_h5 = 'bimcv_metadata.hdf5'
```

```python
df = pd.read_csv(covid19_pandas)
```

```python
count = 20
images = []
labels = []

with h5py.File((covid19_h5), "r") as f:
    fh5_image = f['images']
    for i in range(count):
        images.append(fh5_image[i])
        labels.append(df[df['h5_idx'] == 0]['Position_DL'][0])
```

```python
# Set number of rows and columns
columns = 3
rows = math.ceil(len(images)/columns)
fig = plt.figure(figsize=(15, 5*rows))  # inch 

# ax enables access to manipulate each of subplots
ax = []

# Use local variable to receive subolot
for index, img in enumerate(images):
#     sub_data = COVID_data[i]
#     img = mpimg.imread('./' + sub_data[folder_index] + '/' + sub_data[filename_index])
#     create subplot and append to ax
#     Equals to fig.add_subplot(rows columns i+1)
    subplot = fig.add_subplot(rows, columns, index+1)
    subplot.set_title(f"{index}: {labels[index]}")
    ax.append(subplot)
    plt.imshow(img)

plt.show()
```
